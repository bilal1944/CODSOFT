**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Loading the DataSet**

```python
df = pd.read_csv('/content/creditcard.csv')
df
```

```python
df.shape
```

```python
df.describe()
```

**Checking the Null Values**

```python
df.isnull().sum()
```

**Dropping Null Values**


```python
df.dropna(subset=["Amount"], inplace=True)
df.dropna(subset=["Class"], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
df['Class'].value_counts()
```

```python
normal=df[df.Class==0]
```

```python
fraud=df[df.Class==1]
```

```python
print(normal.shape)
```

```python
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

```python
normal_sample=normal.sample(n=492)
```

```python
new_file=pd.concat([normal_sample,fraud],axis=0)
```

```python
new_file.head(10)
```

```python
new_file['Class'].value_counts()
```

```python
new_file.groupby('Class').mean()
```

**Giving the DataSet to the Model**

```python
X=new_file.drop(columns='Class',axis=1)
```

```python
Y=new_file['Class']
```

**Splitting the DataSet**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

**Model Training**

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

**Train Data Prediction**

```python
X_pred=model.predict(X_train)
```

```python
train_data_acuracy=accuracy_score(X_pred,Y_train)*100
```

```python
print(f"Training Data Accuracy: {train_data_acuracy}%")
```

**Test Data Prediction**

```python
X_test_pred=model.predict(X_test)
```

```python
test_data_accuracy=accuracy_score(X_test_pred,Y_test)*100
```

```python
print(f"Test Data Accuracy: {test_data_accuracy}%")
```
